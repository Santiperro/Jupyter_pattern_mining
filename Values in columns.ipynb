```python
import pandas as pd
data = pd.read_excel('data.xlsx')
data.head(5)
```

```python
# Количество студентов или транзакций
len(data['STUDENT_ID'].unique()) 
```

```python
# Подсчет кол-ва строк бакалавров и магистров
bachelors = []
masters = []
basket2 = []
for i in range(len(data['STUDENT_ID'])):
    if int(data['END_YEAR'][i]) - int(data['BEGIN_YEAR'][i]) == 3:
        bachelors.append(data['STUDENT_ID'][i])
    elif int(data['END_YEAR'][i]) - int(data['BEGIN_YEAR'][i]) == 1:
        masters.append(data['STUDENT_ID'][i])
    else:
        basker2.append(data['STUDENT_ID'][i])
print(len(bachelors))
print(len(masters))
print(len(basket2))
```

```python
# Подсчёт количества бакалавров и магистров
bachelors = pd.Series(bachelors).unique()
print(len(bachelors))
masters = pd.Series(masters).unique()
print(len(masters))
```

```python
# Убеждение в отсутствии студентов с одним и тем же ID в бакалавриате и магистратуре
bm = list(set(bachelors) & set(masters))
bm
```
